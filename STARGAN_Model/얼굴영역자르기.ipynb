# Train 자르기

```python
import os
import cv2
path_dir = r'C:\Users\YongTaek\Desktop\sample\RaFD\train'
result_dir = r'C:\Users\YongTaek\Desktop\finalimage\train'
file_list = os.listdir(path_dir)
for i in file_list:
    print(i)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

for i in file_list:
    emo_dir = os.path.join(path_dir,i)
    res_emo_dir = os.path.join(result_dir,i)
    file = os.listdir(emo_dir)
    for j in file:
        file_name = os.path.join(emo_dir,j)
        img = cv2.imread(file_name)
        
        if '045' in j :
            img = img[150:740, 80:600]
        elif '090' in j :
            img = img[150:700, 100:580]
        elif '135' in j :
            img = img[150:750, 100:580]
        img = cv2.resize(img, dsize=(256, 256), interpolation=cv2.INTER_AREA)
        cv2.imwrite(os.path.join(res_emo_dir, j),img)
```
